## Interpreting a Text Classifier

One of the biggest challenges is understanding how users feel about the movies they watch. Sure, you can look at the star ratings they give, but those aren't always the most informative. That's why you've decided to dig deeper and use interpretable AI to analyze movie reviews and understand the sentiment behind them.
